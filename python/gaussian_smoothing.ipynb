```python
from numpy import r_, sin, array
from scipy.signal import cspline1d, cspline1d_eval
%pylab inline

x = r_[0:10]
dx = x[1]-x[0]
newx = r_[-3:13:0.1]  # notice outside the original domain 
y = array([1,1,1,1,1,0,0,0,1,0], dtype='float')
cj = cspline1d(y)
newy = cspline1d_eval(cj, newx)
from pylab import plot, show
plot(newx, newy, x, y, 'o')
show()
```
