```python
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('multipage.pdf')
import matplotlib.pyplot as plt

plt.plot([1,2,20], '.--')
plt.text(0.05,0.95,"normal plot", transform=gcf().transFigure, size=20)
pp.savefig()

plt.figure()
plt.text(0.05,0.95,"pcolor plot", transform=gcf().transFigure, size=20)
plt.pcolor(array([1,2,3]), array([5,6,7]), array([[1,10],[5,2]]))
pp.savefig()
pp.close()
```
